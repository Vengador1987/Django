## Navegador y estilos

Ya con la parte anterior finalizada, vamos a agreagr una barra de nagevación y agregar más estilos a nuestra app **`blog`**.

Primero, debemos tener los archivos **`navegador.html`**, **`main.html`**, **`main.css`** y **`article.html`**.

Vamos copiar el contenido de **`navegador.html`** en **`base.html`** justo donde comienza la etiqueta de **`<body>`**, este código describe una barra de navegación.

Del archivo **`main.html`** vamos a tomar todo el código y reemplazar este trozo del archivo **`base.html`**, es decir:
Reemplazar esto: 

```html
<div class = "container">
    {% block contenido %}{% endblock %}
</div>
```
Por esto:

```html
<main role="main" class="container">
  <div class="row">
    <div class="col-md-8">
      {% block content %}{% endblock %}
    </div>
    <div class="col-md-4">
      <div class="content-section">
        <h3>Our Sidebar</h3>
        <p class="text-muted">You can put any information here you'd like.
          </p><ul class="list-group">
            <li class="list-group-item list-group-item-light">Latest Posts</li>
            <li class="list-group-item list-group-item-light">Announcements</li>
            <li class="list-group-item list-group-item-light">Calendars</li>
            <li class="list-group-item list-group-item-light">etc</li>
          </ul>
        <p></p>
      </div>
    </div>
  </div>
</main>
```

Para el archivo **`main.css`** vamos a necesitar un directorio llamado **`static`**, ubicada dentro de la carpeta de nuestra app **`blog`**, luego creamos otra carpeta llamada **`blog`** (igual que el nombre de la app). En esta carpeta (**`blog`**) tendremos archivos **`.css`** y **`.js`**.

Ahora para "enlazar" **`main.css`** con **`base.html`**:
- En la primera linea del código hacemos referencia a este directorio (**`static`**) usando **`codeblock`**:
```html
{% load static %}
```

- Antes del **`codeblock`** del **`if/else`** agregamos:
```html
<link rel="stylesheet" type="text/css" href="{% static 'blog/main.css' %}">
```

    - Usando **`{% static 'blog/main.css' %}`** asignamos la ruta del archivo **`main.css`** a la parte de **`href`**.
    
Con el archivo **`article.html`**, vamos a copiar su contenido y reemplazar del archivo **`home.html`** el bloque de código dentro del bucle **`for`**, es decir:
Cambiar esto:
```html
<h1> {{ post.file }} </h1>
<p>Hecho por {{ post.author }} el {{ post.date_posted }}</p>
<p>{{ post.content }}</p>
```
Por esto:
```html
<article class="media content-section">
  <div class="media-body">
    <div class="article-metadata">
      <a class="mr-2" href="#">{{ post.author }}</a>
      <small class="text-muted">{{ post.date_posted }}</small>
    </div>
    <h2><a class="article-title" href="#">{{ post.title }}</a></h2>
    <p class="article-content">{{ post.content }}</p>
  </div>
</article>
```

**Ejecuta otra vez el servidor para ver los cambios.**


Ya tenemos algo más presentable, sin embargo algunos enlaces de nuestra app web no funcionan, principalmente porque no hemos creado esas páginas, pero de las que si tenemos parecen funcionar, esto es porque en el archivo **`home.html`** hemos escrito la ruta de estas vistas, pero lo hemos hecho haciendo **hard coding**.

Para hacer esto más dinámico debemos cambiar la parte de **`href`** de algunas lineas:

```html
<a class="navbar-brand mr-4" href="/">Django Blog</a>
<a class="nav-item nav-link" href="/">Home</a>
<a class="nav-item nav-link" href="/about">About</a>
```
Por:
```html
<a class="navbar-brand mr-4" href="{% url 'blog-home' %}">Django Blog</a>
<a class="nav-item nav-link" href="{% url 'blog-home' %}">Home</a>
<a class="nav-item nav-link" href="{% url 'blog-about' %}">About</a>
```

Usando estos **`codeblock`** vemos que hacemos referencia a los nombres que les dimos a las funciones **`path`** en el archivo **`url.py`** de la app **`blog`**, ahora cada vez que quisieramos cambiar la ruta de esa vista, no tendremos que modificar este archivo porque ahora es dinámico.

**Ejecuta otra vez el servidor para ver los cambios.**
